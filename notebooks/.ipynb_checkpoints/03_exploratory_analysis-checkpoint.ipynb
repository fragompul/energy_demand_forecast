```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('../data/clean/energy_spain_clean.csv', parse_dates=['date'])
```

```python
# Evolución total y renovables
plt.figure(figsize=(12,6))
plt.plot(df['date'], df['electricity_generation'], label='Total')
plt.plot(df['date'], df['renewables_electricity'], label='Renovables')
plt.plot(df['date'], df['fossil_electricity'], label='Fósiles')
plt.title('Generación Eléctrica en España')
plt.xlabel('Año')
plt.ylabel('GWh')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Heatmap correlaciones
plt.figure(figsize=(12,10))
sns.heatmap(df.drop(columns=['year', 'date']).corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Mapa de Correlación Variables Energéticas')
plt.show()
```

```python
# Scatter plots para relaciones clave
sns.pairplot(df[['electricity_generation', 'renewables_electricity', 'fossil_electricity', 'gdp', 'population']])
plt.suptitle('Relaciones entre variables clave', y=1.02)
plt.show()
```
